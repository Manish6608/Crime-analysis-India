```python
import matplotlib.pyplot as plt

selected_columns = ['Rape', 'Kidnapping and Abduction', 'Dowry Deaths', 'Assault on women with intent to outrage her modesty', 'Insult to modesty of Women', 'Cruelty by Husband or his Relatives', 'Importation of Girls']
data_heatmap = data[selected_columns]

plt.matshow(data_heatmap.corr())
plt.title('Heatmap of Selected Crime Types')
plt.xticks(range(len(selected_columns)), selected_columns, rotation=45)
plt.yticks(range(len(selected_columns)), selected_columns)
plt.show()
```

```python
india_dowry_deaths = data[(data['Crime Type'] == 'Dowry Deaths') & (data['Year'] >= 2001) & (data['Year'] <= 2012)]
india_dowry_deaths_by_year = india_dowry_deaths.groupby('Year').size()

plt.bar(india_dowry_deaths_by_year.index, india_dowry_deaths_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Dowry Deaths')
plt.title('Dowry Deaths in India (2001-2012)')
plt.show()
```
